```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"
```

```python
city_info = pd.read_csv(city_data)
ride_info = pd.read_csv(ride_data)

city_info = city_info.drop_duplicates('city')
city_info.head()
```

```python
combined_data = pd.merge(city_info, ride_info, on="city", how='outer')
combined_data.head()
```

```python
city_type = combined_data.groupby("type")["fare"].sum().reset_index()
city_type
```

## Total Fares by City Type

```python
colors = ['Gold', 'SkyBlue', 'Coral']
fig1, ax1 = plt.subplots()
ax1.pie(city_type['fare'], explode=(0,0,0.1), labels = city_type['type'], autopct='%1.1f%%', colors=colors, shadow=True, startangle=160)
ax1.axis('equal')
plt.title('% of Total Fares by City Type', fontsize=12)
plt.savefig('% of Total Fares by City Type.png')
plt.show()
```

## Total Rides by City Type

```python
# Total Rides by City Type
total_rides = combined_data.groupby("type")["ride_id"].count().reset_index()
total_rides
```

```python
colors = ['Gold', 'SkyBlue', 'Coral']

fig1, ax1 = plt.subplots()
ax1.pie(total_rides['ride_id'], explode=(0,0,0.1), labels = total_rides['type'], autopct='%1.1f%%', colors=colors, shadow=True, startangle=90)
ax1.axis('equal')
plt.title('% of Total Rides by City Type', fontsize=12)
plt.savefig('% of Total Rides by City Type.png')
plt.show()
```

## Total Drivers by City Type

```python
#Total Drivers by City Type
total_drivers = combined_data.groupby('type')["driver_count"].sum().reset_index()
total_drivers
```

```python
colors = ['Gold', 'SkyBlue', 'Coral']

fig1, ax1 = plt.subplots()
ax1.pie(total_drivers["driver_count"], explode=(0,0,0.1), labels = total_drivers['type'], autopct='%1.1f%%', colors=colors, shadow=True, startangle=140)
ax1.axis('equal')
plt.title('% of Total Drivers by City Type', fontsize=12)
plt.savefig('% of Total Drivers by City Type.png')
plt.show()
```

## Bubble Plot

```python
updated_city = city_info.set_index("city")
```

```python
#Average Fares/City
average_city = combined_data.groupby("city")["fare"].mean()
#Total Rides per City
totalrides_city = combined_data.groupby("city")["ride_id"].count()
#Total Drivers/city
total_drivers = combined_data.groupby("city")["driver_count"].mean()

city_type = updated_city["type"]
```

```python
 pyber_df = pd.DataFrame({"Average Fare per City": average_city,
                        "Total Rides per City": totalrides_city,
                    "Total Drivers per City": total_drivers,
                          "City Type": city_type})
```

```python
pyber_df.head()
```

```python
#City Type
Rural_df =  pyber_df.loc[pyber_df["City Type"]== "Rural"]
Suburban_df =  pyber_df.loc[pyber_df["City Type"]== "Suburban"]
Urban_df =  pyber_df.loc[pyber_df["City Type"]== "Urban"]
```

```python
areas = pyber_df.groupby("City Type")["Total Drivers per City"].size()
areas
```

```python
ax1 = Urban_df.plot(kind='scatter', x='Total Rides per City', y='Average Fare per City', marker="o", s=areas*20, color='Coral', edgecolor='black', alpha=1, label="Urban", figsize=(12,8))
ax2 = Suburban_df.plot(kind='scatter', x='Total Rides per City', y='Average Fare per City', marker="o", s=areas*10, color='SkyBlue', edgecolor='black', alpha=1, ax=ax1, label='Suburban')
ax3 = Rural_df.plot(kind='scatter', x='Total Rides per City', y='Average Fare per City', marker="o", s=areas*5, color='Gold', edgecolor='black', alpha=1, ax=ax1, label='Rural')


plt.text(.95, 0.5, "Note: Circle size corresponds with the driver count per city.", fontsize=15, transform=plt.gcf().transFigure)
plt.legend(fontsize=20)
plt.grid()
plt.title('Pyber Ride Sharing Data (2016)', fontsize=20)
plt.xlabel('Total Number of Rides (Per City)', fontsize=20)
plt.ylabel('Average Fare ($)', fontsize=20)

plt.savefig('Pyber Ride Sharing Data.png')
```

# Observations
    1. Urban cities accounted for over 62% of total fares. 
    2. Rural areas had some of the highest "Average Fares" amongst the                different city types (with the least amount of total rides).
    3. Suburban areas made up about a third of the total fares.
